```python
#Ambil ulang dataset
url='https://raw.githubusercontent.com/asepmuhidin/MK-Data-Mining-UPB/refs/heads/main/dataset/titanic.csv'
df=pd.read_csv(url, index_col='PassengerId')
df.head()
```

```python
df.isna().sum()
```

```python
#Karena kolom Cabin terlalu besar yang kosong(>50%), maka hapus saja kolomnya
df.drop('Cabin', axis=1, inplace=True)
df.head()
```

```python
#Untuk data Age, tidak dihapus karena sedikit data kosong. Lebih baik lakukan imputasi
#Untuk data numerik bisa di impute dengan nilai rata-rata, median atau modus
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.isna().sum()
```

```python
#Untuk kolom kategorik contohnya kolom Embarked, maka inputasi dengan data yang paling banyak muncul
#cek dahulu data kategori Embarked beserta jumlahnya
df['Embarked'].value_counts()
```

```python
#Kategori 'S' paling banyak muncul, maka 'S' ini yang dipakai untuk imputasi
df['Embarked'].fillna('S', inplace=True)
df.isna().sum()
```

```python
df.describe()
```

```python
#1. Berapa jumlah penumpang yang berumur di atas 29
#2. Berapa penumpang yang tidak bayar(tidak membeli tiket)?

jumlah_penumpang = df[df["Age"] > 29].shape[0]
print(f"Jumlah penumpang umur di atas 29 tahun:{jumlah_penumpang} dari total {df.shape[0]}")
```

```python
penumpang_tidak_bayar = df[df["Fare"] == 0]
jumlah_penumpang_tidak_bayar = penumpang_tidak_bayar.shape[0]
print("Jumlah penumpang yang tidak membayar tiket:", jumlah_penumpang_tidak_bayar)
print(penumpang_tidak_bayar['Name'])
```
